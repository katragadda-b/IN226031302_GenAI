**Answers**

1. "Love" and "love" can be interpreted in 2 different ways in nlp.



*   Case sensitive - if we haven't performed case folding - process of converting text to lowercase/upper case , then "Love" and "love" are interpreted as 2 different values as they each have different ascii values
*  Case insensitive - if we have performed case folding , then the input is considered same as the ascii values of the letters remains the same

2. Stop words are words in a sentence/statement which add zero context to the input. These words are known to increase "noise" in the input , which afect the accuracy. Apart from this they cause high dimensionality storage problem due to redundant data. Due to this , the processing time increases by multifold , causing the model to work slow

3. Removing stopwords which tend to flip the context of a statement - no , never , switch , but etc during sentiment analysis can cause harm as removing these words will change the whole meaning , making the model think the opposite of whatever its intended to.

Eg: let sentence be "the product is great but its not worth the money" will be converted to "product great worth money" which completely switches the whole meaning into something positive despite the original statement not being a positive review

4. Lemmatizing and stemming are two ways to process input data.

stemming removed prefixes and suffixes which could convert an existing word to a non existing word which maynot hold a meaning. Lemmatiizing removes prefizes and suffixes in a way that it actually leads to an existing word in the english dictionary which holds a meaning

```python
"""
Test your function on at least 10 diverse sentences, including:
Emojis
Numbers
Slang
Repeated characters
URLs
Mixed case
Sample Inputs
[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

"""


import re

#Using similar code as task 2

def preprocess(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Reduce repeated characters (3+ → 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove punctuation / emojis
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Token filtering
    tokens = text.split()
    tokens = [t for t in tokens if len(t) > 2 or t in {"no", "not"}]

    return tokens


sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens = preprocess(s)
    cleaned_sentence = " ".join(tokens)

    print("Original Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print()
```

```python

"""
Task 4: Token Analytics
For each processed sentence, compute:
Total number of tokens
Number of unique tokens
Average token length
Analysis Questions
Which sentence had the most noise?
Which sentence retained the most meaningful tokens after cleaning?

"""
results = []
for s in sentences:
    tokens = preprocess(s)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    results.append((s, tokens, total_tokens, unique_tokens, avg_length))

    print("Sentence:", s)
    print("Tokens:", tokens)
    print("Total tokens:", total_tokens)
    print("Unique tokens:", unique_tokens)
    print("Average token length:", round(avg_length, 2))
    print()
```

```python
"""
Task 5: Frequency Analysis
Combine all tokens and use:
from collections import Counter

Identify:
Top 10 most frequent words
Top 5 least frequent words
"""

from collections import Counter

all_tokens = []

for s in sentences:
    all_tokens.extend(preprocess(s))

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
least_common = counter.most_common()[::-1][:5]
print(least_common)
```

```python
"""

Task 6: Build Full Pipeline
def full_pipeline(text_list):
    pass

Expected Output
{
    "tokens": [...],
    "clean_sentences": [...]
}

"""
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
"""
Task 7: Error Handling
Ensure your solution handles:
Empty string
Only emojis
Only numbers

"""
def preprocess(text):
    import re

    if not text or not isinstance(text, str):
        return []

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Reduce repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove punctuation / emojis
    text = re.sub(r'[^\w\s]', '', text)

    # Normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()

    if text == "":
        return []

    tokens = text.split()
    tokens = [t for t in tokens if len(t) > 2 or t in {"no", "not"}]

    return tokens
```
